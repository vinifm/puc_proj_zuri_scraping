```python
# Puxando stopwords do Natural Language ToolKit (NLTK)
nltk.download('stopwords')
stopAll = stopwords.words('portuguese')
```

```python
def get_page(url):
    response = requests.get(url)
    content = response.content
    soup = bs(content, 'html.parser')
    return soup.get_text()

def clean_content(pages: list) -> list:
    cleaned_content = []
    for content in pages:
        content = re.sub(r'\s+', ' ', content)

        # remove pontuação
        punctuation = string.punctuation + "-–"
        content = ''.join(char for char in content if char not in punctuation)
        cleaned_content.append(content)

    return cleaned_content

# Adiciona uma coluna no dataframe com o conteúdo da página
def get_pages_from_dict(url_df: pd.DataFrame) -> pd.DataFrame:
    results = map(get_page, list(url_df['url']))
    cleaned_content = clean_content(list(results))
    url_df['content'] = cleaned_content

    words = []
    for page in url_df['content']:
        words.append([word.lower() for word in page.split() if word not in stopAll])
    url_df['words'] = words

    return url_df
```

```python
df = get_pages_from_dict(url_df)
df
```

### Criando um RDD
Um RDD (Resilient Distributed Dataset) no PySpark é a estrutura de dados fundamental do Apache Spark, representando uma coleção distribuída de elementos que podem ser processados em paralelo. Ele é projetado para ser imutável e tolerante a falhas, permitindo operações de transformação (como map e filter) e ações (como count e collect) de forma eficiente e distribuída. Os RDDs podem ser criados a partir de dados presentes no sistema de arquivos local, HDFS, ou outras fontes de dados, e são fundamentais para o processamento e análise de grandes volumes de dados em ambientes distribuíd

```python
df_word_count = pd.DataFrame()
site_list = []
word_count_list = []
for index, row in grouped_df.iterrows():
    rdd = sc.parallelize(row['words'])
    counts = rdd.map(lambda word:(word, 1)).reduceByKey(lambda a, b: a + b)
    count_list = sorted(counts.collect(), key=lambda x: x[1], reverse=True)

    # filtra somente palavras que ocorrem mais de duas vezes
    filtered_count = [item for item in count_list if item[1] > 2]
    word_count_list.append(filtered_count)
    site_list.append(row['site'])

df_word_count['site'] = site_list
df_word_count['word_count'] = word_count_list

df_word_count
```

Para verificar a ocorrência e frequência de termos específicos nos textos coletados e analisar sua relevância para anúncios de produtos voltados para o público-alvo

```python
# Lista de termos específicos
termos_especificos = [
    'base', 'base líquida', 'pó', 'corretivo', 'pele negra', 'tom', 'autoestima', 'bem estar'
]

# Termos positivos e negativos
termos_positivos = ['autoestima', 'bem estar', 'confiança', 'beleza', 'saudável', 'luminosa']
termos_negativos = ['problema', 'dificuldade', 'ruim', 'imperfeito', 'decepção', 'caro', 'errado', 'tom errado']

# Função para contar a frequência de termos específicos
def contar_termos(texto, termos):
    contagem = {termo: 0 for termo in termos}
    texto = texto.lower()  # Convertendo para minúsculas para padronização
    for termo in termos:
        # Usando regex para garantir que a frase exata seja contada
        if termo in texto:
            contagem[termo] = len(re.findall(r'\b' + re.escape(termo) + r'\b', texto))
    return contagem

# Função para processar os dados e contar termos específicos
def processar_termos_especificos(url_df, termos):
    url_df = get_pages_from_dict(url_df)
    contagem_termos = []
    for index, row in url_df.iterrows():
        texto = row['content']
        contagem = contar_termos(texto, termos)
        contagem['site'] = row['site']
        contagem_termos.append(contagem)
    return pd.DataFrame(contagem_termos)

# Processar termos específicos nos textos
df_termos_especificos = processar_termos_especificos(url_df, termos_especificos + termos_positivos + termos_negativos)
df_termos_especificos.head()
```

```python
# Exibir a frequência dos termos específicos por site
df_frequencia_termos = df_termos_especificos.groupby('site').sum()

# Calcular a relevância dos termos para anúncios
def calcular_relevancia(df, termos_positivos, termos_negativos):
    df['relevancia_positiva'] = df[termos_positivos].sum(axis=1)
    df['relevancia_negativa'] = df[termos_negativos].sum(axis=1)
    return df

df_relevancia = calcular_relevancia(df_termos_especificos, termos_positivos, termos_negativos)
df_relevancia[['site', 'relevancia_positiva', 'relevancia_negativa']].head()
```

Para realizar uma análise de sentimento e identificar palavras ou termos que aparecem com alta frequência em um contexto negativo

```python
# Função para identificar termos negativos frequentes por página
def termos_negativos_frequentes_por_pagina(df, termos_negativos):
    termos_neg_freq = {}
    for termo in termos_negativos:
        termos_neg_freq[termo] = df[termo]
    termos_neg_freq_df = pd.DataFrame(termos_neg_freq)
    return termos_neg_freq_df

# Exibir a frequência dos termos negativos por página
termos_neg_freq_df = termos_negativos_frequentes_por_pagina(df_termos_especificos, termos_negativos)
termos_neg_freq_df
```

Análise de sentimento

```python
# Função para coletar e limpar as páginas, e contar as palavras
def get_pages_from_dict(url_df: pd.DataFrame) -> pd.DataFrame:
    results = map(get_page, list(url_df['url']))
    cleaned_content = clean_content(list(results))
    url_df['content'] = cleaned_content

    words = []
    for page in url_df['content']:
        words.append([word.lower() for word in page.split() if word not in stopAll])
    url_df['words'] = words

    return url_df

# Coletar e limpar os textos
df_pages = get_pages_from_dict(url_df)
```

```python
# Função para analisar o sentimento
def analisar_sentimento(texto):
    nltk.download('vader_lexicon') # Download the lexicon for sentiment analysis
    sentiment_analyzer = SentimentIntensityAnalyzer()
    score = sentiment_analyzer.polarity_scores(texto)
    return score['compound']

# Adicionar coluna de sentimento ao DataFrame
df_pages['sentimento'] = df_pages['content'].apply(analisar_sentimento)
df_pages.head()
```

```python
# Agregar resultados de sentimento por site
df_sentimento_agrupado = df_pages.groupby('site')['sentimento'].mean().reset_index()
df_sentimento_agrupado.columns = ['site', 'sentimento_medio']

# Exibir os resultados
print(df_sentimento_agrupado)
```
